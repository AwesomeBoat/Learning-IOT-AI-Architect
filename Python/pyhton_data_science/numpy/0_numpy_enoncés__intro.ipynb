# Exercice 1
- créez un numpy array depuis la liste suivante [11.7, 20, 325, 10.2, 11.7, 20]
- remplacez la première occurence de la valeur 11.7 par 5
- remplacez toutes les occurences des valeurs 20 par 0
- calculez la taille (mémoire) du tableau
- modifier le typage en uint8
- calculez la taille du tableau
- Observez les valeurs ? Que s'est-il passé ?

<br>

`````python
import numpy as np 

arr1 = np.array([11.7, 20, 325, 10.2, 11.7, 20])
arr1[0] = 5

# OU
arr1 = np.array([11.7, 20, 325, 10.2, 11.7, 20])
arr1_bool_mask = arr1 == 11.7

arr1_modify_first_occur = arr1[arr1_bool_mask]
arr1_modify_first_occur[0] = 5

print(arr1_modify_first_occur)

print(arr1)

#  OU

arr_1 = arr1 = np.array([11.7, 20, 325, 10.2, 11.7, 20])
arr1[np.where(arr1==11.7)[0][0]] = 5

display("ay {}".format(arr1))

# utiliser le masquage booléan OU np.where(cond, if true, if false)
arr1 = np.where(arr1==20, arr1*0, arr1)
display(arr1)
print("taille mémoire : ",arr1.nbytes)
display(arr1.size)
arr1  = arr1.astype(np.uint8)
print("taille mémoire : ",arr1.nbytes)

# La taille mémoire est 8x plus petite (elle doit être en 64 par défaut)
`````

# Exercice 2

- créez un numpy array depuis la matrice suivante [[1, 2, 3],  [4, 5, 6], [7, 8, np.nan]]
- affichez:
  - le type de données de la matrice
  - sa forme
  - son nombre de dimension
  - son nombre d'élément
  - Depuis la forme, pouvez-vous obtenir le nombre de dimension et le nombre d'élément ?
- multipliez la dernière colonne par 5 et divisez la première ligne par 0
- modifier le typage en int. Que se passe-t-il ?

<br>

`````python
arr2 = np.array([[1, 2, 3],  [4, 5, 6], [7, 8, np.nan]])
display(arr2)
print("type de donnée :",arr2.dtype)
print("forme de l'array : ", arr2.shape)
print("nbre dimension : ",arr2.ndim)
print("nbre éléments: ", arr2.size)
print(f"depuis la forme, nbre de dimension = {len(arr2.shape)} et nbre d'éléments = {arr2.shape[0]*arr2.shape[1]}")
arr2[:,-1] *= 5  # Mutiplier la dernière colonne par 5

arr2[0] /= 0 # Diviser la première ligne par 0

print(arr2)

arr2= arr2.astype(np.int32) # Lorsque je transforme le type de l'array en int, les infinites et nan deviennes des nbres négatifs

arr2



`````
