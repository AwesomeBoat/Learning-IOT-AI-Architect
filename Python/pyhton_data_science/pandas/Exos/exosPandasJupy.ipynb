```python
# 21. Quels sont les joueurs ayant gagné des médailles d'Or en badminton simple au cours des années ? Trier le résultat par ordre alphabétique sur le nom des joueurs. 
badminton = (df[df["Sport"]=="Badminton"])
badminton_single_medals = badminton[df["Event"].str.contains("Singles")]
badminton_single_medals = badminton[badminton["Medal"].isin(["Gold","Silver","Bronze"]) ]
badminton_single_medals
```

```python
# 22. Quels sont les 3 pays ayant gagné le plus de médailles de 1984 à 2016 ? 

# Filtrer les athletes medialles
medailles = df[df["Medal"].notna()]

# garder 1 athlète par event par pays et par année
medailles = medailles[["Year","Event","NOC","Medal"]].drop_duplicates()
display(medailles.sort_values(by="Event"))

# Grouper les pays en comptant le nbre de medailles
medailles = medailles.groupby(by="NOC")["Medal"].count().rename("Nb_Medailles").sort_values()
medailles
```
